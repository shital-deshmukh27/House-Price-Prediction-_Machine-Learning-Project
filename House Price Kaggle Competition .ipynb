```python
##Outlier detection using Inter Quantile Range
Q1=data2.quantile(0.25)
Q3=data2.quantile(0.75)
IQR=Q3-Q1
IQR
```

```python
print("outlier Counter of the all features")
((data2 < (Q1 - 1.5 * IQR)) | (data2 > (Q3 + 1.5 * IQR))).sum()
```

```python
##Outliers replace  using Mean 
for col in data2.columns:
    if data2[col].dtypes != 'object':
        q1 , q3 =data2[col].quantile(0.25),data2[col].quantile(0.75)
        iqr = q3 - q1
        ll = q1-1.5*iqr
        ul = q3 + 1.5*iqr
        data2[col] = np.where(data2[col]>ul,data2[col].mean(),np.where(data2[col]<ll,data2[col].mean(),data2[col]))
```
